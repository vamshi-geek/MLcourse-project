```python
def display_menu(mall_name, options):
    print(f"\n{'-'*120}")
    print(f"                 WELCOME TO {mall_name}                ")
    print(f"{'-'*120}")
    print("\nNOTE: DISCOUNT WILL BE APPLIED ONLY AFTER CHOOSING THE PRODUCT\n")
    for option in options:
        print(option)

def calculate_cost(price, discount, quantity):
    discount_amount = quantity * (price * discount)
    total_cost = quantity * price * (1 - discount)
    return total_cost, discount_amount

def display_bill(item, quantity, total_cost, discount_amount):
    print(f"\nOriginal Amount of 1 {item} = {total_cost / quantity:.2f}")
    print(f"Enter number of items you want: {quantity}")
    print(f"Discount Amount = {discount_amount:.2f}")
    print(f"Total Cost = {total_cost:.2f}")

def main():
    total_costs = []

    while True:
        print("\nMAIN MENU:")
        print("PRESS 1: To Order from DUBAI MALL")
        print("PRESS 2: To Order from GOLDEN RESOURCE MALL")
        print("PRESS 3: To Order from MALL OF EMIRATES")
        print("PRESS 4: To Order from THE MALL OF AMERICA")
        print("PRESS 5: To Order from SM MEGA MALL")

        mall_selection = int(input("Make your main menu selection: "))

        if mall_selection == 1:
            display_menu("DUBAI MALL", [
                "PRESS 1: To Order KURTA PYJAMA [RS.1500] (20% Discount)",
                "PRESS 2: To Order JEANS [RS.1000] (10% Discount)",
                "PRESS 3: To Order DENIM JACKETS [RS.2000]",
                "PRESS 4: To Order BLAZERS [RS.1500] (20% Discount)"
            ])
            product_selection = int(input("\nMake your sub-menu selection: "))
            if product_selection == 1:
                price = 1500
                discount = 0.20
                item = "KURTA PYJAMA"
            elif product_selection == 2:
                price = 1000
                discount = 0.10
                item = "JEANS"
            elif product_selection == 3:
                price = 2000
                discount = 0.00
                item = "DENIM JACKETS"
            elif product_selection == 4:
                price = 1500
                discount = 0.20
                item = "BLAZERS"
            else:
                print("\nSORRY......Wrong Selection.....please select among listed options.....\n")
                continue
        elif mall_selection == 2:
            display_menu("GOLDEN RESOURCE MALL", [
                "PRESS 1: To Order KURTA PYJAMA [RS.1200]",
                "PRESS 2: To Order HOODIE [RS.900] (10% Discount)",
                "PRESS 3: To Order POLO SHIRT [RS.1400] (15% Discount)",
                "PRESS 4: To Order DENIM JACKETS [RS.1600] (10% Discount)"
            ])
            product_selection = int(input("\nMake your sub-menu selection: "))
            if product_selection == 1:
                price = 1200
                discount = 0.00
                item = "KURTA PYJAMA"
            elif product_selection == 2:
                price = 900
                discount = 0.10
                item = "HOODIE"
            elif product_selection == 3:
                price = 1400
                discount = 0.15
                item = "POLO SHIRT"
            elif product_selection == 4:
                price = 1600
                discount = 0.10
                item = "DENIM JACKETS"
            else:
                print("\nSORRY......Wrong Selection.....please select among listed options.....\n")
                continue
        elif mall_selection == 3:
            display_menu("MALL OF EMIRATES", [
                "PRESS 1: To Order FLANNEL SHIRT [RS.999] (5% Discount)",
                "PRESS 2: To Order T-SHIRTS [RS.350] (No Discount)",
                "PRESS 3: To Order TOPCOATS [RS.666] (9% Discount)",
                "PRESS 4: To Order BLAZERS [RS.1999] (15% Discount)"
            ])
            product_selection = int(input("\nMake your sub-menu selection: "))
            if product_selection == 1:
                price = 999
                discount = 0.05
                item = "FLANNEL SHIRT"
            elif product_selection == 2:
                price = 350
                discount = 0.00
                item = "T-SHIRTS"
            elif product_selection == 3:
                price = 666
                discount = 0.09
                item = "TOPCOATS"
            elif product_selection == 4:
                price = 1999
                discount = 0.15
                item = "BLAZERS"
            else:
                print("\nSORRY......Wrong Selection.....please select among listed options.....\n")
                continue
        elif mall_selection == 4:
            display_menu("THE MALL OF AMERICA", [
                "PRESS 1: To Order TRENCH COAT [RS.1999] (25% Discount)",
                "PRESS 2: To Order BLAZER [RS.1700] (10% Discount)",
                "PRESS 3: To Order DENIM JACKETS [RS.999]",
                "PRESS 4: To Order POLO SHIRT [RS.1250]"
            ])
            product_selection = int(input("\nMake your sub-menu selection: "))
            if product_selection == 1:
                price = 1999
                discount = 0.25
                item = "TRENCH COAT"
            elif product_selection == 2:
                price = 1700
                discount = 0.10
                item = "BLAZER"
            elif product_selection == 3:
                price = 999
                discount = 0.00
                item = "DENIM JACKETS"
            elif product_selection == 4:
                price = 1250
                discount = 0.00
                item = "POLO SHIRT"
            else:
                print("\nSORRY......Wrong Selection.....please select among listed options.....\n")
                continue
        elif mall_selection == 5:
            display_menu("SM MEGA MALL", [
                "PRESS 1: To Order SWEATER [RS.899] (10% Discount)",
                "PRESS 2: To Order BEANIES [RS.399]",
                "PRESS 3: To Order DENIM JACKETS [RS.2759] (15% Discount)",
                "PRESS 4: To Order HAWAIIAN SHIRT [RS.559]"
            ])
            product_selection = int(input("\nMake your sub-menu selection: "))
            if product_selection == 1:
                price = 899
                discount = 0.10
                item = "SWEATER"
            elif product_selection == 2:
                price = 399
                discount = 0.00
                item = "BEANIES"
            elif product_selection == 3:
                price = 2759
                discount = 0.15
                item = "DENIM JACKETS"
            elif product_selection == 4:
                price = 559
                discount = 0.00
                item = "HAWAIIAN SHIRT"
            else:
                print("\nSORRY......Wrong Selection.....please select among listed options.....\n")
                continue
        else:
            print("\n.....ENTERED INCORRECT OPTION.....\n")
            continue

        quantity = float(input(f"\nEnter number of {item} you want: "))
        total_cost, discount_amount = calculate_cost(price, discount, quantity)
        display_bill(item, quantity, total_cost, discount_amount)
        total_costs.append(total_cost)

        continue_selection = int(input("\nIF YOU WANT TO CONTINUE PRESS 1 \nELSE ENTER ANY OTHER KEY TO SHOW THE BILL: "))
        if continue_selection != 1:
            break

    total_bill = sum(total_costs)
    for idx, cost in enumerate(total_costs, start=1):
        print(f"\nBill Amount for order {idx}: {cost:.2f}")
    print(f"\nTotal Bill Amount for all ordered items: {total_bill:.2f}")
    print("\nTHANK YOU FOR USING OUR PRIME CORNER ONLINE CLOTH STORE. HAVE A NICE DAY....VISIT AGAIN...")

if __name__ == "__main__":
    main()
```
